```python
# dictionary to store matching accuracy each sequence, each camera
matching_accuracy = {}
```

```python
import matplotlib.pyplot as plt
seqs = sorted(glob.glob('new_clusters/seq*'))
# print(seqs)

for seq in seqs[:1]:
    print(os.path.basename(seq).upper())
    matching_accuracy[os.path.basename(seq)] = {}
    cameras = sorted(glob.glob(os.path.join(seq, 'cam*')))
    for cam in cameras:
        count = 0
        same = 0
        people = sorted(glob.glob(os.path.join(cam, 'person*')))
        # print(people[0])
        for person in people:
            print(os.path.basename(person))
            # ids = print(glob.glob(os.path.join(person, '*.webp')))
            ids = [os.path.basename(f).split('_')[-2] for f in glob.glob(os.path.join(person, '*.webp'))]
            ids = set(ids)
            # print(ids)
            for id in ids:
                frames = sorted(glob.glob(os.path.join(person, f'*_{id}_*.webp')))
                # print(f'Person: {person}, ID: {id}, Frames: {frames}')

            # for each id, pick random 1 frame, and display the image
            frames_to_show = []
            for id in ids:
                frames = sorted(glob.glob(os.path.join(person, f'*_{id}_*.webp')))
                frames_to_show.append(frames[0]) 
            if len(frames_to_show) == 1:
                same += 1
                count += 1
                continue

            fig, axs = plt.subplots(1, len(frames_to_show), figsize=(3, 2))
            for i, frame in enumerate(frames_to_show):
                img = plt.imread(frame)
                axs[i].imshow(img)
                axs[i].set_title(f'ID: {os.path.basename(frame).split("_")[-2]}')
                axs[i].axis('off')
            plt.show()    
            temp = int(input("Same person? (1 yes / 2 no): "))
            if temp == 1:  
                same += 1
            count += 1
            accuracy = same / count if count > 0 else 0
            matching_accuracy[os.path.basename(seq)][os.path.basename(cam)] = accuracy
            plt.close()
    print(matching_accuracy[os.path.basename(seq)])
```

## ID Switch

```python
benchmark_dir = 'benchmark/reid/reappearance/train'

predict_dir = 'old_data/metadata_objects'

ID_SWITCHES = {}

# print(glob.glob(f'{benchmark_dir}/train_*/seq_*/imgs/camera*'))

for f in glob.glob(f'{benchmark_dir}/train_*/seq_*/imgs/camera*'):
    # print(f)
    # print('_'.join(os.path.basename(f).split('_')[:2]))  
    camera = '_'.join(os.path.basename(f).split('_')[:2])
    # print('seq:', f.split('/')[-3])
    seq = f.split('/')[-3]
    files = os.listdir(f)
    number_of_gt_ids = len(files)
    print('Number of GT IDs:', number_of_gt_ids)

    metadata_file = f'{predict_dir}/{seq}/{seq}_{camera}.txt'
    with open(metadata_file, 'r') as file:
        lines = file.readlines()
        predict_ids = set()
        for line in lines:
            parts = line.strip().split(' ')
            predict_id = parts[3]
            predict_ids.add(predict_id)
        print(predict_ids)
        number_of_predict_ids = len(predict_ids)
        print(number_of_predict_ids)
    print('Average ID Switches per GT ID:', number_of_predict_ids / number_of_gt_ids)
    ID_SWITCHES[f'{seq}_{camera}'] = number_of_predict_ids / number_of_gt_ids
```

```python
sum = 0.0
for value in ID_SWITCHES.values():
    sum += value
print('Average ID Switches:', sum / len(ID_SWITCHES.values()))
```
